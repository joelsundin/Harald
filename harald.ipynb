```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

model_name = "AI-Sweden-Models/gpt-sw3-6.7b-v2"

tokenizer = AutoTokenizer.from_pretrained(model_name)

# Load quantized if possible. If model supports 4-bit or 8-bit, use that.
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    load_in_8bit=True,   # or load_in_4bit=True if supported
    torch_dtype=torch.float16,   # helps reduce memory
    trust_remote_code=True
)
```

```python
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    device_map="auto",
    trust_remote_code=True
)

prompt = "'Det ar validigt svar for skriver ett hundra svenska ords.' Is this correct?"

output = generator(
    prompt,
    max_new_tokens=100,
    do_sample=True,
    temperature=0.7,
    top_p=0.9
)

print(output[0]["generated_text"])
```
