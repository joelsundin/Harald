```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

device = "cuda:0" if torch.cuda.is_available() else "cpu"
model_id = "bigscience/bloom-3b"
HF_TOKEN = ""

tokenizer = AutoTokenizer.from_pretrained(model_id, use_auth_token=HF_TOKEN)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    torch_dtype=torch.float16,
    use_auth_token=HF_TOKEN
)
```

```python
history = f"{system_prompt}\n"

def chat(user_input, history):
    if is_safe(user_input):
        history += f"User: {user_input}\nAssistant:"
        inputs = tokenizer(history, return_tensors="pt").to(model.device)
        outputs = model.generate(
            **inputs,
            max_new_tokens=150,
            pad_token_id=tokenizer.eos_token_id,
            do_sample=True,
            temperature=0.7
        )
        reply = tokenizer.decode(outputs[0], skip_special_tokens=True)
        # text after last "Assistant:"
        reply = reply.split("Assistant:")[-1].strip()
        history += f" {reply}\n"
        return reply, history
    else:
        return "Unsafe message detected.", history

# Interaktiv loop
print("Chatta med Harald! (Skriv 'exit' för att avsluta)\n")
while True:
    user_input = input("Du: ")
    if user_input.lower() in ["exit", "quit"]:
        print("Avslutar chatten. Hej då! 👋")
        break
    reply, history = chat(user_input, history)
    print(f"Harald: {reply}\n")
```
